```python
import matplotlib.pyplot as plt
import pydicom
from skimage import io
```

```python
# Load the images (example using DICOM format)
picture1 = pydicom.dcmread(r'C:\Users\Favour\Downloads\archive (1)\ST000001\SE000002\IM000001.dcm')
picture2 = pydicom.dcmread(r'C:\Users\Favour\Downloads\archive (1)\ST000001\SE000001\IM000015.dcm')
```

```python
pic1 = picture1.pixel_array
pic2 = picture2.pixel_array
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(pic1, cmap='gray')
plt.title('Picture 1')

plt.subplot(1, 2, 2)
plt.imshow(pic2, cmap='gray')
plt.title('Picture 2')

plt.show()
```

```python
import numpy as np
import cv2  # OpenCV can be used for resizing
```

```python
# Resize img2 to match the dimensions of img1
pic2_resized = cv2.resize(pic2, (pic1.shape[1], pic1.shape[0]))
```

```python
# Fuse images using averaging
fused_image = (pic1 + pic2_resized) / 2
```

```python
# Display the fused image
plt.imshow(fused_image, cmap='gray')
plt.title('Fused Image')
plt.show()
```

```python
from numpy.fft import fft2, fftshift
```

```python
# Apply Fourier Transform
fft_pic1 = fftshift(fft2(pic1))
fft_pic2 = fftshift(fft2(pic2))
```

```python
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(np.log(np.abs(fft_pic1)), cmap='gray')
plt.title('Fourier Transform of Picture 1')

plt.subplot(1, 2, 2)
plt.imshow(np.log(np.abs(fft_pic2)), cmap='gray')
plt.title('Fourier Transform of Picture 2')

plt.show()
```

```python
from skimage.filters import sobel
```

```python
# Apply Sobel filter for edge enhancement
edges1 = sobel(pic1)
edges2 = sobel(pic2)# Apply Sobel filter for edge enhancement
```

```python
# Display the edge-enhanced images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(edges1, cmap='gray')
plt.title('Edge Enhanced Photo 1')
```

```python
plt.subplot(1, 2, 2)
plt.imshow(edges2, cmap='gray')
plt.title('Edge Enhanced Photo 2')

plt.show()
```

```python
from skimage.transform import rotate
import matplotlib.pyplot as plt
```

```python
# Rotate the images by 45 degrees
rotated_pic1 = rotate(pic1, angle=45)
rotated_pic2 = rotate(pic2, angle=45)
```

```python
# Display the rotated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(rotated_pic1, cmap='gray')
plt.title('Rotated Photo 1 at 45 degrees')

plt.subplot(1, 2, 2)
plt.imshow(rotated_pic2, cmap='gray')
plt.title('Rotated Photo 2 at 45 degrees')

plt.show()
```
